```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor


color_pal = ["#F8766D", "#D39200", "#93AA00", "#00BA38", "#00C19F", "#00B9E3", "#619CFF", "#DB72FB"]
```

```python
df = pd.read_csv('/Users/mikeyling/Desktop/Jupyter Notebooks/Cheezy Money/TSLA.csv', index_col='Date', parse_dates=True)
```

```python
df['20d_roll_avg'] = df['Close'].rolling(20).mean()
df['20d_std'] = df['Close'].rolling(20).std()
df['upper_bollinger'] = df['20d_roll_avg'] + 2 * df['20d_std']
df['lower_bollinger'] = df['20d_roll_avg'] - 2 * df['20d_std']
df['buy_bollinger']

df.dropna(inplace=True)
```

```python
df.head(45)
```

```python
tmp_df = df

fig,ax = plt.subplots(figsize=(15,6))
plt.plot(tmp_df.index, tmp_df['upper_bollinger'], color=color_pal[0], label='20D Bollinger')
plt.plot(tmp_df.index, tmp_df['lower_bollinger'], color=color_pal[0])
plt.plot(tmp_df.index, tmp_df['20d_roll_avg'], color=color_pal[2], label='20D SMA')
plt.plot(tmp_df.index, tmp_df['Close'], color='#CCCCFF', label='Closing Price')
plt.legend()

plt.show()
```
